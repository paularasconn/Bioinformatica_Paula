##### 2.- Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
```

```python
# Crear listas:
accs_a = [acc for acc in accs if acc.startswith('a')]
accs_resto = [acc for acc in accs if not acc.startswith('a')]
```

```python
#vemos que efectivamente se han separado correctamente
accs_a
```

```python
accs_resto 
```

```python
#Crear archivos
with open("accs_a.txt", "w") as f_a:
    for acc in accs_a:
        f_a.write(acc + "\n")

with open("accs_rest.txt", "w") as f_rest:
    for acc in accs_resto:
        f_rest.write(acc + "\n")
```

```python
print("Accessiones que empiezan por 'a':", accs_a)
print("Accessiones restantes:", accs_resto)
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

 

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
R1 = [a for a in accs if re.search(r'5', a)]
print("Contienen el número 5:", R1)
```

```python
R2 = [a for a in accs if re.search(r'[de]', a)]
print("Contienen la letra d o la e:", R2)
```

```python
R3 = [a for a in accs if re.search(r'd.*e', a)]
print("Contienen las letras d y e en ese orden:", R3)
```

```python
R4 = [a for a in accs if re.search(r'd.e', a)]
print("Contienen las letras 'd' y 'e' en ese orden con una letra cualquiera en medio:", R4)
```

```python
R5 = [a for a in accs if re.search(r'd', a) and re.search(r'e', a)]
print("Contiene ambas letras 'd' y 'e' en cualquier orden", R5)
```

```python
R6 = [a for a in accs if re.search(r'^[xy]', a)]
print("Empieza por 'x' o por 'y' letras 'd' y 'e' en cualquier orden", R6)
```

```python
R7 = [a for a in accs if re.search(r'^[xy].*e$', a)]
print("Empiezan por 'x' o 'y' y terminan por 'e'", R7)
```

```python
R8 = [a for a in accs if re.search(r'\d{3,}', a)]
print("tienen tres o más dígitos seguidos", R8)
```

```python
R9 = [a for a in accs if re.search(r'd[arp]$', a)]
print("Terminan con 'd' seguidos de 'a' o 'r' o 'p'", R9)
```
